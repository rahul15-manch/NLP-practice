```python
import pandas as pd 
messages = pd.read_csv('/Users/rahulmanchanda/Desktop/NLP/text_representation/SpamClassifier-master/SMSSpamCollection.txt', sep='\t', names=["label", "message"])
```

```python
messages.head()
```

```python
import re
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
from nltk.stem import WordNetLemmatizer
lematizer = WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(messages)):
    review=re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lematizer.lemmatize(word) for word in review if word not in set('english stopwords')]
    review=' '.join(review)
    corpus.append(review)
```

```python
result = [[i, j, k] 
          for i, j, k in zip(map(len, corpus), corpus, messages['message']) 
          if i < 1]
```

```python
result
```

```python
words=[]
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
for sentence in corpus:
    sent_token = sent_tokenize(sentence)  # this breaks the paragraph into sentences
    for sent in sent_token:
        words.append(simple_preprocess(str(sentence)) ) # deacc=True removes punctuations
```
